```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Exercise 1

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

x = np.linspace(0, 100, 50)
y = x * 2

ax.plot(x, y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.set_xlim([min(x), max(x)])
ax.set_ylim([min(y), max(y)])

plt.show()
```

```python
# Exercise 2

fig = plt.figure()

pts = [[0, 0, 1, 1], [0.2, 0.5, 0.2, 0.2]]

for p in pts:
    ax = fig.add_axes(p)
    ax.plot(x, y)
    
    ax.set_title('title')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    
    ax.set_xlim([min(x), max(x)])
    ax.set_ylim([min(y), max(y)])
    
plt.show()
```

```python
# Exercise 3

file_path = 'company_sales_data.csv'
df = pd.read_csv(file_path)

plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')

plt.show()
```

```python
# Exercise 4

file_path = 'company_sales_data.csv'
df = pd.read_csv(file_path)

plt.plot(df['month_number'], df['total_profit'], 'r:')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(['Total Profit'], loc='lower right')

plt.show()
```

```python
plt.plot(df['month_number'], df['total_units'], 'ro:', linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold Units')
plt.legend(['Sold Units'], loc='lower right')

plt.show()
```
