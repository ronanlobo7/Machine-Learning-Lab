```python
class MyPolynomialRegression:
    def __init__(self):
        pass
    
    def fit(self, X, Y):
        n = len(X)
        pow_x = [np.sum(X ** i) for i in range(5)]
        
        print(pow_x)
        
        xiy = np.array([[np.sum(X**i * Y)] for i in range(3)], dtype=np.float32)
        
        print(xiy)
        
        Mat = np.array([[n, pow_x[1], pow_x[2]],
                      [pow_x[1], pow_x[2], pow_x[3]],
                      [pow_x[2], pow_x[3], pow_x[4]]], dtype=np.float32)
        
        print(Mat)
        
        self.coef = np.linalg.inv(Mat).dot(xiy)
        
    def predict(self, X):
        return sum([self.coef[i][0] * X ** i for i in range(3)])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(2)
X_poly = poly_features.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, Y)
```

```python
Y_pred = model.predict(X_poly)
```
