```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('golddata.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
X_label, Y_label = data.columns[0], data.columns[1]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(data[X_label], data[Y_label])
plt.xlabel(X_label)
plt.ylabel(Y_label)
plt.title('Gold Prices by Year')
plt.show()
```

```python
X, Y = data['Year'].values.reshape(-1, 1), data['Gold Prices in Rs'].values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(3)
X_poly = poly_features.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, Y)

Y_pred = model.predict(X_poly)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(data['Year'], data['Gold Prices in Rs'], label='Actual Prices')
plt.plot(data['Year'], Y_pred, '--', label='Predicted Prices')
plt.xlabel(data.columns[0])
plt.ylabel(data.columns[1])
plt.title('Gold Prices by Year')
plt.legend()
plt.show()
```
